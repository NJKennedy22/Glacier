```python
class Ground:
    def __init__(self,C):
        if C ==1:
            pass
        if C==2:
            pass
        if C==3:
            pass
        if C==4:
            pass

class Unit:
    def init_Infantry(self):
        self.Strength=30
        self.AttH=5
        self.AttM=25
        self.AttL=50
        self.DefH=15
        self.DefM=25
        self.DefL=50
        self.Mounteff=[30,30,30]
        self.Swampeff=[-20,-20,-20]
        self.Plainseff=[15,-10,10]
        self.Walleff=[30,30,30]
    def init_Sniper(self):
        self.Strength=15
        self.AttH=10
        self.AttM=20
        self.AttL=100
        self.DefH=10
        self.DefM=10
        self.DefL=20
        self.Mounteff=[30,30,30]
        self.Swampeff=[15,15,-10]
        self.Plainseff=[15,-30,10]
        self.Walleff[30,30,30]
    def init_Jeep(self):
        self.Strength=40
        self.AttH=15
        self.AttM=50
        self.AttL=80
        self.DefH=15
        self.DefM=50
        self.DefL=70
        self.Mounteff=[-10,20,10]
        self.Swampeff=[0,-30,-10]
        self.Plainseff=[15,0,10]
        self.Walleff=[30,30,30]
    def init_Artillery(self):
        self.Strength=50
        self.AttH=90
        self.AttM=80
        self.AttL=70
        self.DefH=15
        self.DefM=30
        self.DefL=40
        self.Mounteff=[10,10,30]
        self.Swampeff=[0,-10,10]
        self.Plainseff=[15,0,10]
        self.Walleff=[30,30,30]
    def init_Chopper(self):
        self.Strength=75
        self.AttH=50
        self.AttM=60
        self.AttL=90
        self.DefH=80
        self.DefM=30
        self.DefL=90
        self.Mounteff=[0,0,0]
        self.Swampeff=[0,0,0]
        self.Plainseff=[0,0,0]
        self.Walleff=[0,0,0]

    def init_smTank(self):
        self.Strength=60
        self.AttH=60
        self.AttM=70
        self.AttL=60
        self.DefH=40
        self.DefM=50
        self.DefL=70
        self.Mounteff=[-10,-10,0]
        self.Swampeff=[0,-10,0]
        self.Plainseff=[20,20,20]
        self.Walleff=[30,30,30]
    def init_bgTank(self):
        self.Strength=90
        self.AttH=80
        self.AttM=100
        self.AttL=70
        self.DefH=60
        self.DefM=70
        self.DefL=90
        self.Mounteff=[-20,-10,-10]
        self.Swampeff=[0,-20,0]
        self.Plainseff=[20,20,20]
        self.Walleff=[30,30,30]
    def __init__(self,C):
        if C == 1:
            self.init_Infantry()
        if C ==2:
            self.init_Sniper()
        if C ==3:
            self.init_Jeep()
        if C==4:
            self.init_Artillery()
        if C==5:
            self.init_Chopper()
        if C==6:
            self.init_smTank()
        if C==7:
            self.init_bgTank()
class Tile:
    def init_unit(self,C):
        self.kind = 0
        self.T = Unit(C)
    def init_ground(self,C):
        self.kind = 1
        self.T = Ground(C)
    def __init__(self,C):
        self.ID = C
        self.Life = 10
        self.path = 0
        self.move_down =0
        self.move_right =0
        self.checked =0
        if C >10:
            self.init_unit(C-10)
        elif C>0:
            self.init_ground(C)
        else:
            self.Life =0
    def change_tile(self,C):
        if C.ID>10:
            self.init_unit(C.ID-10)
        elif C.ID>0:
            self.init_ground(C.ID)
        self.Life = C.Life
        self.ID =C.ID
    
    
        
    
```

```python
def check_empty(T):
    if T.Life ==0 and T.path ==0 and T.move_down ==0 and T.move_right==0 and T.checked ==0:
        return 1
    
    
def check_fallable(T):
    if T.Life !=0 and T.path ==0 and T.move_down==0 and T.move_right==0 and T.checked==0:
        return 1
    
    
def check_above_fall(game_board,row,col):
    flag = 1
    r = row
    while flag and r != 0:
        r-=1
        if not check_empty(game_board[r][col]):
            if check_fallable(game_board[r][col]):
                return 1
            else:
                return 0
    
def check_left_fall(game_board,row,col):
    flag = 1
    c = col
    while flag and c != 0:
        c-=1
        if not check_empty(game_board[row][c]):
            if check_fallable(game_board[row][c]):
                return 1
            else:
                return 0
    
def set_above_fall(game_board,row,col):
    flag = 1
    r =row
    count =1
    while flag and r !=0:
        r-=1
        if check_empty(game_board[r][col]):
            game_board[r][col].path = 1
            count+=1
        else:
            if check_fallable(game_board[r][col]):
                game_board[r][col].move_down = count
            else:
                flag = 0
            
def set_left_fall(game_board,row,col):
    flag = 1
    c =col
    count =1
    while flag and c !=0:
        c-=1
        if check_empty(game_board[row][c]):
            game_board[row][c].path = 1
            count+=1
        else:
            if check_fallable(game_board[row][c]):
                game_board[row][c].move_right = count
            else:
                flag = 0
    
def move_board(game_board):
    rows = len(game_board)
    cols = len(game_board[0])
    for row in range (rows-1, -1,-1):
        for col in range(cols-1, -1,-1):
            if(game_board[row][col].move_down != 0):
                game_board[row + game_board[row][col].move_down][col].change_tile(game_board[row][col])
                game_board[row][col].Life =0
            if(game_board[row][col].move_right != 0):
                game_board[row][col+game_board[row][col].move_right].change_tile(game_board[row][col])
                game_board[row][col].Life =0

def reset_board(game_board):
    rows = len(game_board)
    cols = len(game_board[0])
    for row in range (rows-1, -1,-1):
        for col in range (cols-1, -1,-1):
            game_board[row][col].move_down =0
            game_board[row][col].move_right =0
            game_board[row][col].path =0
            game_board[row][col].checked =0
            if game_board[row][col].Life ==0:
                game_board[row][col].ID =0

    
def settle_board(game_board,rows,cols):
    flag = 1
    while flag==1:
        flag =0
        for row in range (rows-1,-1,-1):
            for col in range (cols-1,-1,-1):
                if check_empty(game_board[row][col]):
                    if check_above_fall(game_board,row,col):
                        set_above_fall(game_board,row,col)
                        flag = 1
                    elif check_left_fall(game_board,row,col):
                        set_left_fall(game_board,row,col)
                        flag = 1
                    else:
                        game_board[row][col].checked = 1
        move_board(game_board)
        reset_board(game_board)
    
                    
                    
```
